```python
import pandas as pd

url = "https://docs.google.com/spreadsheets/d/1NcHu1Ec8C_2XDnjQo_ECb_-r3NtxEm1J9OpDee4MzUs/export?format=csv"

df = pd.read_csv(url)
print(df.head())
```

```python
# Xử lý lại tên cột cho dễ dùng
df.columns = [
    "timestamp",
    "customer_name",
    "phone",
    "booking_time",
    "event_date",
    "note"
]

# Convert kiểu dữ liệu
df["event_date"] = pd.to_datetime(df["event_date"], dayfirst=True, errors="coerce")
df["booking_time"] = pd.to_datetime(df["booking_time"], format="%H:%M:%S", errors="coerce").dt.time

print(df.head())
```
